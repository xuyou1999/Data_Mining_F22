## Convert all the JSON -> CSV

```python
#Start Processing JSON
for number in ["01","02","03","04","05","06","07","08","09","10","11","12","13"]:
    Time = []
    LineID = []
    DirectionID = []
    DistanceFromPoint = []
    PointID = []
    
    
    
    file_path = "../data/vehiclePosition" + number + ".json"
    
    #A DataFrame
    json_file = pd.read_json(file_path,orient="split")
    
    #For all the tuple in the DataFrame
    for i in range(json_file.shape[0]):
        t = json_file["time"].iloc[i]              #Time
        Responses = json_file["Responses"].iloc[i] #All Responses
        
        # Line Index
        if i %500 ==0:
            print(i)
        
        # There is no data in Responses in this line
        if all(Responses) == False:
            continue
        
        #each response
        for j,response in enumerate(Responses):
            lines = response["lines"]
            
            # Line data in this response is empty
            if len(lines) == 0:
                continue
            
            #Each Line
            for line in lines:
                lineid = line['lineId']
                vehiclePositions = line['vehiclePositions']
                
                #Each Vehicle in this line
                for position in vehiclePositions:
                    directionid = position['directionId']
                    distancefrompoint = position['distanceFromPoint']
                    pointid = position['pointId']
                    
                    Time.append(t)
                    LineID.append(lineid)
                    DirectionID.append(directionid)
                    DistanceFromPoint.append(distancefrompoint)
                    PointID.append(pointid)
                    
    df = pd.DataFrame({'Time':Time,
                       'LineID':LineID,
                       'DirectionID':DirectionID,
                       'DistanceFromPoint':DistanceFromPoint,
                       'PointID':PointID})
    
    print("For Loop Done")
    #print(Time[0])
    #print(LineID[0])
    #print(DirectionID[0])
    #print(DistanceFromPoint[0])
    #print(PointID[0])
    
    print(df.head(5))
    print(df.shape)
    
    #Convert it into CSV
    df.to_csv("../data/vehiclePosition" + number + ".csv",index=False)
    print(number +" Done!")
    print("------------------------------------------------")
                    
                    
                
                
        
        
        
```
